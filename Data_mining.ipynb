Web Scraping / Data mining

```python
# import packages

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
link='https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG'
page = requests.get(link) #  Send HTTP request to the URL of the page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
soup.find('div', attrs={'class':'_4rR01T'})
```

```python
products=[]
prices=[]
ratings=[]

for data in soup.findAll('div',class_='_3pLy-c row'):
        names=data.find('div', attrs={'class':'_4rR01T'})
        price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})

        products.append(names.text)
        prices.append(price.text)
        ratings.append(rating.text)

print(len(products))
print(len(ratings))
print(len(prices))
print(products)
print(ratings)
print(prices)
```

```python
df=pd.DataFrame({'Product Name':products,'Price':prices,'Rating':ratings})
df
```

```python
df.to_csv('new.csv', index=False) # index=False to avoid saving the DataFrame index as a column
```
